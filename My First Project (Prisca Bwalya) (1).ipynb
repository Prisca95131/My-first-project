# THE DATA BELOW EXPLAINS THE COLUMS.
Order_ID ......unique identifier for each order, groups multiple products within the same order
Customer_ID....Unique identifier for each customer , distinguishing individual buyers
Customer_Type..Indicates whether the customer is B2B (Business to Business) or B2C (Business to Consumer)
Product........The name of the product purchased suc as "Coca-Cola" or "Erdinger WeiBbier"
Category.......The product category such as "soft drinks" or "Alcoholic Beverages"
Unit_Price.....The price per unit of the product.
Quantity.......The number of units purchased for the specified product in the order
Discount.......The discount applied to the product (eg., 0.1 for 10%). Discounts are only given to B2B customers
Total_Price....The total price of the product after applying discount.
Region.........The Region of the customer such as "Bayern" or "Berlin"
Order_Date.....The date when the order was placed.

QUESTION 2
What is the relationship between quantity and discounts?

Question 3 How much Revenue was lost due to discounts given?

```python
#Expected Revenue without discounts
data_new["Expected_Revenue"] = data_new["Unit_Price"] * data_new["Quantity"]

#Loss due to discounts
data_new["Discount_Loss"] = data_new["Expected_Revenue"] - data_new["Total_Price"]

#total loss from discounts
total_discount_loss = data_new["Discount_Loss"].sum()
```

```python
total_discount_loss
```

```python
plt.figure()

plt.plot(data_new["Expected_Revenue"], label="Revenue Without Discounts")
plt.plot(df["Total_Price"], label="Revenue After Discounts")

plt.xlabel("Sales Transactions")
plt.ylabel("Revenue")
plt.title("Impact of Discounts on Revenue")

plt.legend()
plt.show()
```

Since the gap between the Revenue without Discounts and the Revenue After Discounts is very small, meaning the discounts did not significantly reduce the revenue.
